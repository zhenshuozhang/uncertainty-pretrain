```python
# 导入库
import torch
import matplotlib.pyplot as plt

# 定义数据
A = range(5)
B1 = torch.tensor([1, 3, 5, 7, 9])
B2 = torch.tensor([2, 4, 6, 8, 10])
C = torch.tensor([50, 40, 20, 2, 60])

# 创建图形
fig, ax1 = plt.subplots()
ax1.set_xlabel('X轴')
ax1.set_ylabel('Y轴', color='red')
ax1.plot(B1, color='red', label='A和B')
#ax1.tick_params(axis='y', labelcolor='red')
ax1.plot(B2, color='red', label='A和B')
#ax1.tick_params(axis='y', labelcolor='red')

# 添加双轴
ax2 = ax1.twinx()
ax2.set_ylabel('Y轴', color='blue')
ax2.plot(A, C, color='blue', label='C')
ax2.tick_params(axis='y', labelcolor='blue')

# 添加图例
fig.legend(loc='upper left')

# 显示图形
plt.show()
```
